
# 1. **What is the difference between a function and a method in Python?**

# * **Function:** A block of reusable code that can be called independently. It is defined using `def` outside any class.
# * **Method:** A function that is associated with an object or class; it operates on the object and is called using dot notation.

# **Example:**

# ```python
# # Function
# def greet(name):
#     return f"Hello, {name}!"
# print(greet('Ajay'))  # Hello, Ajay!

# # Method
# s = "python"
# print(s.upper())  # UPPERCASE method of string object
# ```

# 2. **Explain the concept of function arguments and parameters in Python.**

# * **Parameter:** Variable defined in the function definition to accept values.
# * **Argument:** Actual value passed to the function when calling it.

# **Example:**

# ```python
# def add(a, b):  # a, b are parameters
#     return a + b

# result = add(2, 3)  # 2 and 3 are arguments
# print(result)  # 5
# ```

# ---

# 3. **What are the different ways to define and call a function in Python?**

# * **Normal function:** Using `def` keyword.
# * **Lambda function:** One-line anonymous function.
# * **Calling:** Directly by name with parentheses and arguments.

# **Example:**

# ```python
# # Normal function
# def square(x):
#     return x**2
# print(square(4))  # 16

# # Lambda function
# sq = lambda x: x**2
# print(sq(5))  # 25
# ```

# 4. **What is the purpose of the `return` statement in a Python function?**

# * It sends back a value from the function to the caller. Without `return`, a function returns `None`.

# **Example:**

# ```python
# def multiply(a, b):
#     return a*b

# print(multiply(3, 4))  # 12
# ```

# 5. **What are iterators in Python and how do they differ from iterables?**

# * **Iterable:** Any object capable of returning its elements one by one (list, tuple, string).
# * **Iterator:** An object that produces elements of an iterable **on demand** using `__next__()`.

# **Example:**

# ```python
# lst = [1,2,3]  # iterable
# it = iter(lst)  # iterator
# print(next(it))  # 1
# print(next(it))  # 2
# ```

# ---

# 6. **Explain the concept of generators in Python and how they are defined.**

# * Generators are **special functions** that **yield** values one by one using the `yield` keyword. They save memory because they produce items on demand.

# **Example:**

# ```python
# def gen_numbers(n):
#     for i in range(n):
#         yield i

# for num in gen_numbers(3):
#     print(num)  # 0, 1, 2
# ```

# ---

# 7. **What are the advantages of using generators over regular functions?**

# * **Memory efficient:** No need to store entire sequence.
# * **Lazy evaluation:** Values generated on-the-fly.
# * **Infinite sequences possible:** Unlike lists which are finite.

# **Example:**

# ```python
# def infinite_nums():
#     n = 0
#     while True:
#         yield n
#         n += 1

# # We can safely get first 5 numbers
# gen = infinite_nums()
# for _ in range(5):
#     print(next(gen))  # 0,1,2,3,4
# ```

# ---

# 8. **What is a lambda function in Python and when is it typically used?**

# * Lambda function is a **one-line anonymous function**.
# * Typically used with **map(), filter(), sorted()** or anywhere a small temporary function is needed.

# **Example:**

# ```python
# square = lambda x: x**2
# print(square(6))  # 36
# ```

# ---

# 9. **Explain the purpose and usage of the `map()` function in Python.**

# * Applies a function to **all elements of an iterable** and returns an iterator.

# **Example:**

# ```python
# nums = [1,2,3]
# squared = list(map(lambda x: x**2, nums))
# print(squared)  # [1,4,9]
# ```

# 10. **What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

# | Function | Purpose                                                   | Example                                                            |
# | -------- | --------------------------------------------------------- | ------------------------------------------------------------------ |
# | map()    | Apply a function to all elements of an iterable           | `list(map(lambda x:x*2,[1,2,3])) # [2,4,6]`                        |
# | reduce() | Reduce iterable to a single value using a binary function | `from functools import reduce; reduce(lambda x,y:x+y,[1,2,3]) # 6` |
# | filter() | Select elements based on a condition                      | `list(filter(lambda x:x%2==0,[1,2,3,4])) # [2,4]`                  |
